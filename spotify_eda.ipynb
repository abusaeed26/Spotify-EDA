# Release Year Distribution

```python
plt.hist(df['released_year'], bins=20)
plt.title("Songs Release Year Distribution")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

# Streams Distribution

```python
plt.hist(df['streams'], bins=30)
plt.title("Spotify Streams Distribution")
plt.xlabel("Streams")
plt.ylabel("Songs Count")
plt.show()
```

# Audio Feature Distributions

```python
audio_features = [
    'danceability_%', 'energy_%', 'valence_%',
    'acousticness_%', 'instrumentalness_%',
    'liveness_%', 'speechiness_%'
]

df[audio_features].hist(bins=20, figsize=(14,10))
plt.suptitle("Audio Feature Distributions")
plt.show()
```

# Correlation Analysis

```python
corr = df.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# Danceability vs Energy

```python
sns.scatterplot(data=df, x='danceability_%', y='energy_%')
plt.title("Danceability vs Energy")
plt.show()
```

# Playlists vs Streams

```python
sns.scatterplot(data=df, x='in_spotify_playlists', y='streams')
plt.title("Spotify Playlists vs Streams")
plt.show()
```

# Artist Collaboration Impact

```python
sns.boxplot(x='artist_count', y='streams', data=df)
plt.title("Artist Count vs Streams")
plt.show()
```

# BPM Distribution

```python
sns.histplot(df['bpm'], bins=30)
plt.title("BPM (Tempo) Distribution")
*plt.show()
```

*1. Songs with higher Spotify playlist presence receive more streams*

*2. Danceability and energy are strong indicators of song popularity*

 *3. Most popular songs have moderate BPM (90–140)*
   

 *4. Instrumental and acoustic songs are less common among top hits*


*5. Collaborations increase song reach and streams*
